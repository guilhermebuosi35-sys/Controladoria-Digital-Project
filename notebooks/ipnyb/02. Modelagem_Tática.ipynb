### Taxa de Conversão Geral

Explora a taxa de conversão geral das campanhas

***Interpretação:***

Nota-se que a taxa de conversão geral é extremamente positiva, sendo %87.65, representando um número agressivo.

```sql
WITH cte AS (
SELECT
  SUM (CASE WHEN conversion = 1 THEN 1 ELSE 0 END) / COUNT (*) taxa_conversao
FROM workspace.google_drive.cenario_tatico_marketing_campaign)

SELECT
  ROUND(taxa_conversao * 100,2) AS taxa_conversao_geral
FROM cte;
```

### Taxa de Conversão por Canal de Campanha

Descreve a taxa de conversão agrupado por campanha

| Canal  | Comentário                                                                            |
| ---------------- | ------------------------------------------------------------------------------------- |
| Indicação        | Tráfego proveniente de recomendações, parceiros ou sites terceiros.                   |
| Mídia paga (PPC) | Campanhas de anúncios pagos (ex: Google Ads, Facebook Ads), modelo pay-per-click.     |
| SEO              | Otimização para motores de busca, visitas orgânicas via Google/Bing sem custo direto. |
| E-mail           | Disparos e campanhas realizadas via e-mail marketing (ex: newsletter, promoções).     |
| Mídia social     | Ações em redes sociais (ex: Instagram, Facebook, LinkedIn, Twitter, TikTok).          |

***Interpretação:***

Indicação é a campanha com melhor conversão, seguido por mídia paga, o que pode mostrar um caminho interessante, visto que o trafêgo orgânico se mostra um canal essencial e de melhor custo benefício para a empresa.
Entender qual canal focalizar os seus investimentos em campanhas, é um passo maduro para a maior captação de leads. 

```sql
WITH cte AS (
SELECT
  campaign_channel,
  SUM (CASE WHEN conversion = 1 THEN 1 ELSE 0 END) / COUNT (*) taxa_conversao
FROM workspace.google_drive.cenario_tatico_marketing_campaign
GROUP BY (campaign_channel)
)

SELECT
  CASE WHEN campaign_channel = 'Referral' THEN 'Indicação'
    WHEN campaign_channel = 'PPC' THEN 'Mídia Paga (PPC)'
    WHEN campaign_channel = 'Email' THEN 'Email'
    WHEN campaign_channel = 'SEO' THEN 'SEO'
    WHEN campaign_channel = 'Social Media' THEN 'Midia Social' END AS campanha,
  ROUND(taxa_conversao * 100,2) AS taxa_conversao
FROM cte
ORDER BY taxa_conversao DESC;
```

### Taxa de Conversão por tipo de campanha

Evidencia a taxa de conversão agrupado pelo tipo de foco da campanha

| Label  | Comentário                                                                          |
| --------------- | ----------------------------------------------------------------------------------- |
| Conversão       | Fase final, voltada à ação/compra/resultado esperado.                               |
| Retenção        | Voltada a manter clientes ativos/engajados (pós-venda, fidelização).                |
| Reconhecimento  | Fase inicial de exposição da marca/produto, gerar conhecimento no público-alvo.     |
| Consideração    | Público já conhece o produto e está avaliando opções, etapa intermediária do funil. |

***Interpretação:***

O tipo de campanha com maior conversão naturalmente foi focado em conversão final, atingindo uma taxa de %93.36, mostrando o momento do funil em que o lead está mais vunerável — Os demais se estiveram com taxas parecdias, sendo importante também para manter fidelização, rechonhecimento e consideração.

```sql
WITH cte AS (
SELECT
  campaign_type,
  SUM (CASE WHEN conversion = 1 THEN 1 ELSE 0 END) / COUNT (*) taxa_conversao
FROM workspace.google_drive.cenario_tatico_marketing_campaign
GROUP BY (campaign_type)
)

SELECT
  CASE WHEN campaign_type = 'Conversion' THEN 'Conversão'
    WHEN campaign_type = 'Retention' THEN 'Retenção'
    WHEN campaign_type = 'Awareness' THEN 'Reconhecimento/Consciência'
    WHEN campaign_type = 'Consideration' THEN 'Consideração'
    END AS campanha,
  ROUND(taxa_conversao * 100,2) AS taxa_conversao
FROM cte
ORDER BY taxa_conversao DESC;
```

### CAC (Custo por conversão do Cliente)

Analisa o CAC por cada campanha, ou seja, o valor que cada campanha teve de investimento por lead convertido 

***Interpretação:***

Mídia Paga é o tráfego com melhor Custo por conversão, mostrando um bom targeting do fluxo, todavia, o canal de e-mail está gerando um custo alto para pouco retorno, sendo um possível ponto de melhoria na segmentação.

```sql
SELECT 
  CASE WHEN campaign_channel = 'Referral' THEN 'Indicação'
    WHEN campaign_channel = 'PPC' THEN 'Mídia Paga (PPC)'
    WHEN campaign_channel = 'Email' THEN 'Email'
    WHEN campaign_channel = 'SEO' THEN 'SEO'
    WHEN campaign_channel = 'Social Media' THEN 'Midia Social' END AS canal_campanha, 
  ROUND(SUM (ad_spend),2) AS Investimento,
  COUNT_IF (conversion = 1) AS Conversoes,
  ROUND(SUM (ad_spend) / COUNT_IF (conversion = 1),2) AS CAC
FROM workspace.google_drive.cenario_tatico_marketing_campaign
GROUP BY canal_campanha
ORDER BY CAC
;
```

### Média/Mediana sobre o tempo do lead no site

Retorna o tempo médio e mediana dos leads da campanha dentro do site

***Interpretação:***

A média (7,73 min) e a mediana (7,68 min) do tempo no site são bastante próximas, indicando distribuição equilibrada do comportamento dos usuários. Isso sugere que a maioria das visitas segue um padrão semelhante de engajamento, sem grandes variações extremas.

```sql
SELECT ROUND(AVG (time_on_site),2) AS media_tempo_no_site,
  ROUND(PERCENTILE_CONT (0.5) WITHIN GROUP (ORDER BY time_on_site),2) AS mediana_tempo_no_site
FROM workspace.google_drive.cenario_tatico_marketing_campaign;
```

### Média/Mediana de Compartilhamento nas Redes - Canal de prospecção

Retorna a Média/Mediana de compartlhamento nas redes sociais de acordo com cada canal de prospecção

***Interpretação:***

A média e mediana giram em torno de 48-52 compartilhamento nas redes sociais, sem diferença grande dentre os números, sendo a maior em mídia social (Média = 50.85 e Mediana = 52), o que demonstra que temos alguns outliers puxando a média para baixo, ou seja, alguns casos mais significativos de pessoas que não compartilharam o conteúdo — o que é natural.

No demais, o número é alto, e mostra engajamento grande nas redes pelo os leads.

```sql
SELECT 
  CASE WHEN campaign_channel = 'Referral' THEN 'Indicação'
    WHEN campaign_channel = 'PPC' THEN 'Mídia Paga (PPC)'
    WHEN campaign_channel = 'Email' THEN 'Email'
    WHEN campaign_channel = 'SEO' THEN 'SEO'
    WHEN campaign_channel = 'Social Media' THEN 'Midia Social' END AS campanha,
  ROUND(AVG (social_shares),2) AS media_compartilhamento,
  ROUND(PERCENTILE_CONT (0.5) WITHIN GROUP (ORDER BY social_shares),2) AS mediana_compartlhamento
FROM workspace.google_drive.cenario_tatico_marketing_campaign
GROUP BY campanha
ORDER BY media_compartilhamento DESC;
```

### Leads por Faixa etária

Seleciona a quantidade de leads **convertidos** por faixa etária dentro das campanhas

***Interpretação:***

A faixa de 40-50 anos é a segunda maior contagem de convertidos, com maior taxa de conversão, se mostrando um público importante e confiável.

```sql
WITH cte AS (  
  SELECT 
    CASE WHEN age BETWEEN 60 AND 70 THEN '60-70'
      WHEN age BETWEEN 50 AND 60 THEN '50-60'
      WHEN age BETWEEN 40 AND 50 THEN '40-50'
      WHEN age BETWEEN 30 AND 40 THEN '30-40'
      ELSE '18-30' END AS faixa_etaria,
    COUNT_IF (conversion = 1) AS contagem,
    ROUND (SUM(conversion) / COUNT (*) , 2) AS taxa_conversao
  FROM workspace.google_drive.cenario_tatico_marketing_campaign
  GROUP BY faixa_etaria
  ORDER BY faixa_etaria ASC
)

SELECT
  ROW_NUMBER () OVER (ORDER BY faixa_etaria) AS ordem,
  *
FROM cte;
```

```sql
SELECT 
  age,
  COUNT_IF (conversion = 1) AS contagem,
  ROUND (SUM(conversion) / COUNT (*) , 2) AS taxa_conversao
FROM workspace.google_drive.cenario_tatico_marketing_campaign
WHERE age BETWEEN 58 AND 69
GROUP BY age
ORDER BY age ASC
;
```

### Leads por Faixa de Renda

Seleciona a quantidade de leads **convertidos** por faixa de renda dentro das campanhas

***Interpretação:***

Taxa de conversão por faixa de renda é igual, ou seja, a renda dos leads não impacta muito sobre as conversões.   

```sql
WITH agrupamento AS (SELECT income,
  conversion,
  NTILE(5) OVER (ORDER BY income DESC) AS grupos
FROM workspace.google_drive.cenario_tatico_marketing_campaign
ORDER BY income DESC
)

SELECT
  ROW_NUMBER () OVER (ORDER BY grupos DESC) AS ordem,
  'R$ ' || MIN (income) ||' - R$ ' || MAX (income) AS faixa_renda_anual,
  COUNT_IF (conversion = 1) AS contagem,
  ROUND (SUM (conversion) / COUNT (*), 2) AS taxa_conversao 
FROM agrupamento 
GROUP BY grupos;
```
